```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.pylab as pylab
params = {'legend.fontsize': 'x-large',
          'figure.figsize': (8, 5),
         'axes.labelsize': 'x-large',
         'axes.titlesize':'x-large',
         'xtick.labelsize':'x-large',
         'ytick.labelsize':'x-large'}
pylab.rcParams.update(params)

```

```python
seqgan_loss = pd.read_csv("./results/seqgan/experiment-log-seqgan-real.csv")
maligan_loss = pd.read_csv("./results/maligan/experiment-log-maligan-real.csv")
leakgan_loss = pd.read_csv("./results/leakgan/experiment-log-leakgan-real.csv")
losses = {'SeqGAN': seqgan_loss, 'MaliGAN': maligan_loss, 'LeakGAN': leakgan_loss}
```

```python
def plot_result(losses):
    legends = []
    for gan_type, each_loss in losses.items():
        legends.append(gan_type)
        plt.plot(each_loss['epochs'], each_loss['nll-test'])
    plt.title('Model Loss')
    plt.ylabel('NLL')
    plt.xlabel('Epoch')
    plt.legend(legends, loc='upper right')
    plt.show()
```

```python
plot_result(losses)
```
